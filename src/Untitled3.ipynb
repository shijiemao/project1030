```julia


A = [1 2;3 4]
Σ = A*A'
μ = 1
d = Distributions.AbstractMvNormal(μ, Σ)
c + B * d    # == MvNormal(B * μ + c, B * Σ * B')
dot(b, d)    # == Normal(dot(b, μ), b' * Σ * b)
```

```julia
using Random
using Statistics
using Plots
Random.seed!(11)

x = rand(Normal(0,1),100)
y = 1 .+ 1/2 .* (x .^2)
m = cov(x,y)/var(x)
b = mean(y)-cov(x,y)/mean(x)
xy = [(x[i],y[i]) for i in 1:100]
scatter(xy)

xs = -2:0.01:2
plot!(xs, b .+ m .*xs,
      linewidth = 1)

```
