```julia
using Distributions
using Statistics
using Random
using Plots
Random.seed!(11)
A = rand(0.0:30.0,2,2)
Σ = A*A'

d1 = MvNormal([1.;5.], Σ)


```

```julia
Random.seed!(11)
ran = rand(d1,100)
x = ran[1,:]
y = ran[2,:]

m = cov(x,y)/var(x)
b = mean(y) - cov(x,y)/var(x)*mean(x)

range_h = minimum(y):0.01:maximum(y)
range_p = minimum(x):0.01:maximum(x)

heatmap(range_p,range_h, (x,y)->pdf(d1,[x,y]))
scatter!(x,y, pointsize = 1, color = "blue", legend = false)

plot!(x, m .*x .+ b)
```
