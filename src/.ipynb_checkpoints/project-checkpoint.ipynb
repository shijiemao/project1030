```python
Final = []

kf = StratifiedKFold(n_splits=5,shuffle=True,random_state=random_state)
k = 1

for train_index, val_index in kf.split(X_other,y_other):
    X_train = X_other.iloc[train_index]
    y_train = y_other.iloc[train_index]
    X_val = X_other.iloc[val_index]
    y_val = y_other.iloc[val_index]
    print('train balance:')
    print(y_train.value_counts(normalize=True))
    print('val balance:')
    print(y_val.value_counts(normalize=True))
    print("------------preprocessing--------------------fold:"+str(k))
    X_train_tran = clf.fit_transform(X_train)
    X_val_tran = clf.transform(X_val)
    Final.append([X_train_tran, X_val_tran])
    k+=1
```

```python
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import ParameterGrid
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, MinMaxScaler, LabelEncoder

def MLpipe_KFold_F1(X, y, preprocessor, ML_algo, param_grid):
    '''
    This function splits the data to other/test (80/20) and then applies KFold with 4 folds to other.
    The RMSE is minimized in cross-validation.
    '''
    best_models = []
    train_ratio = 0.8
    K = 4 
    random.seed(11) # to make the model building process reproducable. 
    random_s = [random.randrange(1,100) for i in range(5)]
    test_scores = np.zeros(len(random_s))
    test_x = []
    test_y = []
    
    for i in range(len(random_s)):f
        print(i)
        kf = KFold(n_splits=K,shuffle=True,random_state=random_s[i])
        pipe = make_pipeline(preprocessor,ML_algo)
        grid = GridSearchCV(ML_algo, param_grid=param_grid,scoring = 'f1_macro',
                            cv=kf, return_train_score = True, verbose=True)
        X_other, X_test, y_other, y_test = train_test_split(X,y,train_size = train_ratio, random_state= random_s[i])
        grid.fit(X_other, y_other)
    
        print('best model parameters:',grid.best_params_)
        print('validation score:',abs(grid.best_score_)) 
    
        test_x.append(X_test)
        test_y.append(y_test)
        best_models.append(grid)
        y_test_pred = best_models[-1].predict(X_test)
        test_scores[i] =f1_score(y_test, y_test_pred, average = "macro")
        print('test score:',test_scores[i])
        
    return best_models, test_scores

```

```python
from sklearn.ensemble import RandomForestClassifier
def ML_pipeline_kfold(X,y,preprocessor,random_state,n_folds):
    X_other, X_test, y_other, y_test = train_test_split(X, y, test_size=0.2, random_state = random_state)

    kf = StratifiedKFold(n_splits=n_folds,shuffle=True,random_state=random_state)

    pipe = make_pipeline(preprocessor,RandomForestClassifier(random_state = 11))

    param_grid = {'randomforestclassifier__max_depth': [30]}

    grid = GridSearchCV(pipe, param_grid=param_grid,cv=kf, scoring = 'f1_macro', return_train_score = True,n_jobs=-1,verbose=10)

    grid.fit(X_other, y_other)
    return grid, X_test, y_test
```

```python

preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(sparse=False,handle_unknown='ignore'), onehot_ftrs),
        ('std', StandardScaler(), std_ftrs)])

grid, X_test, y_test = ML_pipeline_kfold(features,target,preprocessor,11,4)
test_score = grid.score(X_test,y_test)
print(grid.best_score_)
print(test_score)
ftr_names = features.columns
nr_runs = 10
scores = np.zeros([len(ftr_names),nr_runs])

for i in range(len(ftr_names)):
    print('shuffling '+str(ftr_names[i]))
    acc_scores = []
    for j in range(nr_runs):
        X_test_shuffled = X_test.copy()
        X_test_shuffled[ftr_names[i]] = np.random.permutation(X_test[ftr_names[i]].values)
        acc_scores.append(grid.score(X_test_shuffled,y_test))
    print('   shuffled test score:',np.around(np.mean(acc_scores),3),'+/-',np.around(np.std(acc_scores),3))
    scores[i] = acc_scores
```

```python
sorted_indcs = np.argsort(np.mean(scores,axis=1))[::-1]
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(24,))
plt.boxplot(scores[sorted_indcs].T,labels=ftr_names[sorted_indcs],vert=False)
plt.axvline(test_score,label='test score')
plt.title("Permutation Importances (test set)")
plt.xlabel('score with perturbed feature')
plt.legend()
plt.tight_layout()
plt.savefig('../figures/global importance.png', dpi = 800)
plt.show()
```
