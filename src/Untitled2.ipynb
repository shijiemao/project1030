```julia
using Distributions
using Plots

iter = 10000
i = 0
interval = []
while i<iter
    start = 0
    
    for i in 1:3600
        if rand(Binomial(1, 1/100))==1
            push!(interval, i-start)
            start = i
        end
    end
    i+=1
end
histogram(interval,bar_width = 5, bins = 100, normalize =:density)
```

```julia
using Distributions
using Plots
using StatsBase

iter = 10000
i = 0
intervals = []
while i<iter
    all = sample(1:3600, 36, replace=false)
    sort!(all)
    interval = []
    for k in 1:35

        push!(interval, all[k+1]-all[k])
    end
    for j in interval
        
        push!(intervals, j)
    end
    i+=1
end

histogram(intervals, bar_width = 5, bins = 100,normalize = :density)

```

```julia
using Plots, Distributions
function binom_stickplot(n, p; kw...)
    ν = Binomial(n, p)
    sticks(0:60, k -> pdf(ν, k),
           label = "Binomial($n, $p)"; kw...)
end
binom_stickplot(3600, 1/100, ylims = (0, 0.1))
```
